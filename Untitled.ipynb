# Simulación de Velocidad del Viento Horaria Utilizando Modelos ARIMA y RNN

 Debido a sus bajos costos, bajo impacto ambiental y alta disponibilidad, la energía eólica se ha vuelto una alternativa cada vez más atractiva, en conjunto con la energía solar. El problema asociado a esta energía es que se genera a través del viento, cuya naturaleza variable impone desafíos en su incorporación en los sistemas eléctricos.

##### Definición del problema: 

El problema a ser estudiado en este proyecto es el de construir un modelo que sea capaz de simular una serie de tiempo de la velocidad del viento horaria, con dos propósitos principales:   
-  Utilizar el modelo para simular series de tiempo del viento que permitan realizar estudios asociados a los requerimientos de granjas eólicas (selección de turbinas, gestión de turbinas, etc).
- Utilizar el modelo para realizar predicciones de características futuras de la serie del viento.

Ambos objetivos apuntan a apoyar la integración de la energía eólica en sistemas eléctricos.

##### Conceptualización del modelo:

Se consideran tres opciones para el modelo: 
- Un modelo $sARIMA(p,d,q)\,x\,(P,D,Q)$ que modele las relaciones de la velocidad del viento con sus valores anteriores:

$$\Phi_P(B^s)\, \phi(B) \, (1 - B)^d \,(1 - B^s)^D\, x_t = \Theta_Q(B^s) \, \theta(B) \, w_t $$

- Un modelo $pARMA(p,q)$ con periodo $T$:

$$ X_t = \sum_{j=1}^p \, \phi_j (t) \, X_{t-j} + \sum_{k=1}^q \, \theta_k (t) \, w_{t-k} + \sigma(t) \, w_{t}  $$

En donde $\phi_j (t) = \phi_j (t + T)$, $\theta_k (t) = \theta_k (t + T)$, $\sigma(t) = \sigma(t + T)$, para $j = 1, ... ,p$, $k= 1, ... ,q$ son coeficientes periódicos, y $w_{t}$ es ruido blanco. Este modelo se utiliza cuando $u(t) = u(t + T)$, $j = 1,2, ..., T$ y $ \gamma(t , t + h) = \gamma( t + T , t + T + h) $

- Un modelo de Redes Neuronales Recurrentes basado en la idea de la periodicidad.


$$   \mathbf{s}_{k+1} = f(\mathbf{W_i}\, \mathbf{x}_{k} + \mathbf{W_r}\, \mathbf{s}_t + \mathbf{b}) $$

En donde $\mathbf{x}_{k} = [x_{kT+t}, x_{kT+t-1}, x_{kT+t-2},...,x_{kT+t-T}]^T$ contiene toda la información del periodo.


##### Datos:

Los datos a utilizar corresponden a mediciones promedio de la velocidad del viento cada 10 minutos, medidas por una torre de prospección eólica a 20 metros en la Región de Antofagasta, pertenecientes al Ministerio de Energía, Chile. Las mediciones son promediadas utilizando una media móvil con ventana deslizante de tamaño 6 para obtener el promedio de la velocidad del viento cada 1 hora. En total, se tienen 19536 mediciones.

```python
def input_mask(inputs,units_per_input,units_per_output, prev_activations = 0,initial = False):
    
    #output_dim = activations_per_input*(input_dim-1)      
        
    # si es una capa de entrada, se ignora el último valor del input para lograr la activación.
    # si es una capa intermedia o de salida, el input y el output son iguales.
    # ver cómo manejar el caso de la capa final
    if initial == False:
        
        #assert inputs == outputs
        assert prev_activations == 0
        outputs = inputs
        
    else:
        
        outputs = inputs - 1
        
    mask = np.ones((prev_activations+inputs*units_per_input,outputs*units_per_output))

    for i in range(inputs):
        
        if i == 0:
            
             mask[0:units_per_input+prev_activations,:units_per_output*i] = 0
        
        else:
            
            mask[prev_activations+i*units_per_input:prev_activations+i*units_per_input+units_per_input,:units_per_output*i] = 0 

    return mask
```

```python
from keras import backend as K
from keras.engine.topology import Layer

class MaskedAE(Layer):

    def __init__(self, output_dim, mask, **kwargs):
        self.output_dim = output_dim
        self.mask = mask
        super(MaskedAE, self).__init__(**kwargs)

    def build(self, input_shape):
        # Create a trainable weight variable for this layer.
        self.kernel = self.add_weight(name='kernel', 
                                      shape=(input_shape[1], self.output_dim),
                                      initializer='uniform',
                                      trainable=True)
        super(MaskedAE, self).build(input_shape)  # Be sure to call this at the end

    def call(self, x):
        weights = self.kernel * self.mask
        return K.dot(x, weights)

    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.output_dim)
```

```python
def periodic_predict(n_simulations, horizon, data,periods,model,model_name,path,true_outputs):
    
    '''Parameters
    n_simulations (int): how many times each predictions is going to be sampled
    horizon (int): how many hours ahead the prediction is computed
    data (n_samples,time_steps,lag): wind_data
    periods (n_samples*24): one-hot vector indicating the period corresponding to the sample
    model: trained keras model
    model_name (string): model name
    path (string): something like "results/"
    true_outputs (n_samples,24): true outputs to be used for comparison against predictions
    '''
    
    #periods = to_categorical(np.arange(initial_period, initial_period+simulation_length)%24, num_classes=24)
    x = np.zeros((len(data), horizon))
    predictions = np.zeros((n_simulations,len(data), horizon))
    lag = data.shape[-1]
    time_steps = data.shape[-2]
    current_input = data
    
    for j in range(n_simulations):
        current_input = data
        for i in range(horizon):
                #print("imprimiendo actual")
                #print(current_input)
                #print(periods[:,i,:])
                params = model.predict([current_input,periods[:,i,:].reshape(-1,24)])
                mu, sigma = params[:,0],params[:,1]
                x[:,i] = np.random.normal(mu,sigma)
                #print("mostrando x")
                #print(mu)
                #print(sigma)
                #print(x)
                if time_steps != 1:
                    values = np.concatenate((current_input[:,0,:].reshape(-1,lag),current_input[:,1:,-1].reshape(-1,time_steps-1)),axis=1)
                    values = np.concatenate((values[:,1:], x[:,i].reshape(-1,1)), axis=1)   
                    current_input = np.concatenate([values[:,t:lag + t]  for t in range(time_steps)],axis=1)
                    current_input = current_input.reshape(-1, time_steps, lag)
                elif time_steps == 1 and lag != 1:
                    values = np.concatenate((current_input.reshape(-1,lag)[:,1:],x[:,i].reshape(-1,1)),axis=1)
                    current_input = values.reshape(-1,time_steps,lag)
                   
                else:
                    current_input = x[:,i].reshape(-1,1,1)
        predictions[j,:,:] = x
              
    # fixed 90% confidence interval
    upper_bounds = np.percentile(predictions,5,axis=0)
    lower_bounds = np.percentile(predictions,95,axis=0)
    
    PICP_per_horizon = np.mean((true_outputs>lower_bounds)*(true_outputs<upper_bounds),axis=0)
    PICP = np.mean(PICP_per_horizon)
    
    MPIW_per_horizon = np.mean(upper_bounds - lower_bounds,axis=0)
    MPIW = np.mean(MPIW_per_horizon)
    u = 0.9
    n = 5
    y_per_horizon = PICP_per_horizon < u
    CWC_per_horizon = MPIW_per_horizon*(1+y_per_horizon*np.exp(-n*(PICP_per_horizon-u)))
    y = PICP < u
    CWC = MPIW*(1+y*np.exp(-n*(PICP-u)))
    
    
    return predictions, upper_bounds, lower_bounds, CWC_per_horizon, CWC
```
